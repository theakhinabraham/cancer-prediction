# Import Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

# Data Collection

```python
data = pd.read_csv('data.csv')
```

# Data Cleaning

```python
#Look for nan entries
sns.heatmap(data.isnull())
```

```python
#remove useless columns that do not contribute to our process
data.drop(['Unnamed: 32', 'id'], axis=1, inplace=True)
```

```python
data.head()
```

```python
#Convert Malignant into 1 and Benign into 0
data.diagnosis = [1 if value == "M" else 0 for value in data.diagnosis]
```

```python
data["diagnosis"] = data["diagnosis"].astype("category", copy=False)
```

```python
data["diagnosis"].value_counts().plot(kind="bar")
```

### Split into target values and predictors

```python
# Prepare the model
y = data["diagnosis"] # our target variable
X = data.drop(["diagnosis"], axis=1) # our predictors
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler

# Create a scaler object
scaler = StandardScaler()

# Fit the scaler to the data and transform the data
X_scaled = scaler.fit_transform(X)

# X_scaled is now a numpy array with normalized data
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.30, random_state=42)
```

### Train model

```python
y.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression

# Create logistic regression model
lr = LogisticRegression()

# Train the model on the training data
lr.fit(X_train, y_train)

# Predict the target variable on the test data
y_pred = lr.predict(X_test)
```

```python
y_pred
```

# Evaluate Model

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy: .2f}')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
